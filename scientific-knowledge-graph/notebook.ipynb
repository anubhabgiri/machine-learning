Poc NLP pipeline

clinical free text -> extract entities -> map to CUI

```python
# Starting with a basic example
import spacy
from scispacy.umls_linking import UmlsEntityLinker

nlp = spacy.load("en_core_sci_lg")

nlp.add_pipe(
    "scispacy_linker",
    config={
        "resolve_abbreviations": True,
        "linker_name": "umls"
    }
)

text = """
Patient with history of CHF and diabetes mellitus.
Started on metformin. Blood glucose elevated.
"""

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, ent.label_)
    for cui, score in ent._.kb_ents:
        print("  →", cui, score)

```

```python
# Rule of thumb
# cosine similarity > 0.9 == acceptable
```
