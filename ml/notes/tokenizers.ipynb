```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

sequence = "Using a Transformer network is simple"
tokens = tokenizer.tokenize(sequence)

print("tokens:", tokens)

ids = tokenizer.convert_tokens_to_ids(tokens)

print("input ids:", ids)

decoded_string = tokenizer.decode(ids)
print("decoded string:", decoded_string)

inputs = tokenizer(sequence, padding=True, truncation=True, return_tensors="pt")

print("inputs:", inputs)
```

It is evident that the tensor object generated by the tokenizer is essentially the list of token IDs derived from the original seq plus some other numbers at the beginning and end.
